```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'name': ['Shakil'] * 3 + ['Sejuti'] * 3 + ['Tithi'] * 3,
    'subject': ['Database', 'Data Structure', 'Algorithm'] * 3,
    'score': np.random.randint(60, 100, 9)})
df
```

```python
def get_subject_rank(df):
    new_df = df.copy()
    new_df['subject_rank'] = (new_df.groupby(['subject'])['score'].rank(ascending=False))
    
    return new_df

df.pipe(get_subject_rank)
```

```python
def get_subject_rank(copy, df_or_not=True):
    copy = df.copy()
    if df_or_not is True:
        copy['subject_rank'] = (copy.groupby(['subject'])['score'].rank(ascending=False))
        return copy
    else:
        output_series = (copy.groupby(['subject'])['score'].rank(ascending=False))
        return output_series

df.pipe(get_subject_rank, df_or_not=True)

```

```python
# data is not the first argument

def add_score(duplicate_df, added_score):
    duplicate_df = df.copy()
    duplicate_df = duplicate_df.assign(new_score=lambda x: x.score+added_score)
    return duplicate_df

df.pipe(add_score, 4)
```

```python
candidates = pd.DataFrame({'full_name':['Almirr Shakil','Kawser Mahbub','Jannatun Nuha','Mohosin Kabir','Imran Hossain','Modina Ahmed'],
             'company':['cramstack','Tiger IT','Brainstation 23','micro1','Mediusware','Brainstation 23'],
             'Application_date': ['11/17/2022', '09/23/2022', '12/2/2021', 
                                  '08/25/2022', '01/07/2022', '12/26/2022'],
             'From_office (min)': [120,95,75, 80,100, 34]})
```

```python
candidates['Application_date'] = pd.to_datetime(candidates['Application_date'])
```

```python
def get_first_and_second_name(candidates, col_name):
    copy_df = candidates.copy()
    splitted_full_name = copy_df[col_name].str.split(' ', expand=True)
    
    copy_df['first_name'] = splitted_full_name.get(0)
    copy_df['last_name'] = splitted_full_name.get(1)
    
    return copy_df

first_method = get_first_and_second_name(candidates, 'full_name')
first_method
```

```python
def get_application_date_info(candidates, col_name):
    application_date = df[col_name]
    copy_df = candidates.copy()
    
    copy_df['Day'] = application_date.dt.day
    copy_df['Month'] = application_date.dt.month
    copy_df['Year'] = application_date.dt.year
    copy_df['Day_of_week'] = application_date.dt.day_name()
    copy_df['Month_of_year'] = application_date.dt.month_name()
    
    return copy_df

result = get_application_date_info(candidates, 'Application_date')
result
```
